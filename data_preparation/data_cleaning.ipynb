```python
import pandas as pd
```

```python
data_1 = pd.read_csv('bina_az_19092023.csv')
data_2 = pd.read_csv('bina_az_21092023.csv')
```

```python
data_1
```

```python
data_2
```

```python
frames = pd.concat([data_1, data_2], copy=True,   ignore_index=True,)
frames.tail()
```

```python
frames=frames.drop_duplicates(keep='last')
frames
```

```python
# frames.to_excel('frames.xlsx',index=False)
```

```python
frames.isna().sum()
```

```python
frames_rm_na=frames.dropna(ignore_index=True)
frames_rm_na
```

```python
frames_rm_na.isna().sum()
```

```python
columns=pd.DataFrame(frames_rm_na.columns)
columns
```

```python
main_data=frames_rm_na[['view','title','seller_type','price','currency','location','address','address_all','description','building_type','all_data']]
main_data
```

```python
main_data.info()
```

```python
main_data['update_info'] = main_data['view'].str.extract(r'(?i)(?:Yeniləndi|Обновлено): (.*?),(?:Baxışların sayı|Просмотров):')
main_data['view_count'] = main_data['view'].str.extract(r'(?i)(?:Baxışların sayı|Просмотров): (\d+)')
main_data.drop('view', axis=1, inplace=True)
main_data
```
