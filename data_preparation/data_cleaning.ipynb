```python
import pandas as pd
import numpy as np
```

```python
data_1 = pd.read_csv('bina_az_19092023.csv')
data_2 = pd.read_csv('bina_az_21092023.csv')
```

```python
data_1
```

```python
data_2
```

```python
frames = pd.concat([data_1, data_2], copy=True,   ignore_index=True,)
frames.tail()
```

```python
frames=frames.drop_duplicates(keep='last')
frames
```

```python
# frames.to_excel('frames.xlsx',index=False)
```

```python
frames.isna().sum()
```

```python
frames_rm_na=frames.dropna(ignore_index=True)
frames_rm_na
```

```python
frames_rm_na.isna().sum()
```

```python
columns=pd.DataFrame(frames_rm_na.columns)
columns
```

```python
main_data=frames_rm_na[['view','title','seller_type','price','currency','location','address','address_all','description','building_type','all_data']]
main_data
```

```python
main_data.info()
```

```python
# main_data['update_info'] = main_data['view'].str.extract(r'(?i)(?:Yeniləndi|Обновлено): (.*?),(?:Baxışların sayı|Просмотров):')
main_data['view_count'] = main_data['view'].str.extract(r'(?i)(?:Baxışların sayı|Просмотров): (\d+)')
main_data.drop('view', axis=1, inplace=True)
main_data
```

```python
# main_data.to_excel('main_data.xlsx',index=False)
```

```python
main_data['seller_type'].unique()
```

```python
seller_type_mapping = {
    'vasitəçi (agent)': 0,
    'посредник (агент)': 0,
    'собственник': 1,
    'mülkiyyətçi': 1
}

main_data['seller_type_encoded'] = main_data['seller_type'].replace(seller_type_mapping)
main_data.drop('seller_type', axis=1, inplace=True)

main_data
```

```python
main_data['price']
```

```python
main_data['price'] = main_data['price'].str.replace(' ', '').astype(float)

main_data['price'].sum()
```

```python
main_data = main_data.sort_values(by='price', ascending=True)
main_data
```

```python
main_data.head(3)
```

```python
main_data['description_length'] = main_data['description'].apply(len)
main_data
```

```python
main_data['building_type'].unique()
```

```python
# mapping dictionary to unify similar values
building_type_mapping = {
    'Köhnə tikili': 'Old building',
    'Yeni tikili': 'New building',
    'Ofis': 'Office',
    'Офис': 'Office',
    'Объект': 'Object',
    'Дом / Дача': 'House / Cottage',
    'Вторичка': 'Secondary',
    'Həyət evi / Bağ evi': 'House / Garden house',
    'Новостройка': 'New building',
    'Obyekt': 'Object',
    'Mənzil': 'Apartment',
    'Участок': 'Plot',
    'Torpaq': 'Land',
    'Qaraj': 'Garage',
    'Гараж': 'Garage'
}

main_data['building_type_unified'] = main_data['building_type'].replace(building_type_mapping)
```

```python
main_data
```

```python
main_data['building_type_unified'].unique()
```

```python
main_data['all_data']
```

```python
main_data[['property_type', 'floor_info', 'area', 'room_count', 'data_1', 'data_2', 'data_3']] = main_data['all_data'].str.split(',', expand=True)
```

```python
# new column 'is_near_metro' based on the presence of 'm.' or 'metro' in the specified columns
main_data['is_near_metro'] = (main_data['location'].str.contains('m\.', case=False) |
                               main_data['address_all'].str.contains('m\.', case=False) |
                               main_data['description'].str.contains('m\.', case=False) |
                               main_data['location'].str.contains('metro', case=False) |
                               main_data['address_all'].str.contains('metro', case=False) |
                               main_data['description'].str.contains('metro', case=False)).astype(int)

main_data
```
